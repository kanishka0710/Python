# Tic-Tac-Toe
This is a Python based Tic-Tac-Toe game

The game will accept 1 or 2 players. If you are a solo player, the computer will play against you. 

**Guidelines**
1. Please only input 1-9 for the moves.
2. Please input uppercase X or O for the icon. 
3. Have Fun!


```python
import random

#Initializes board array to save game
board = [" " for i in range(9)]

# Player 1 name
player1 = ""

#Player 2 Name
player2 = ""

# 0 index is player 1 icon
# 1 index is player 2 icon
icon1 = ""
icon2 = ""

```

```python
# Displays the board when called with all updated board values

def display_board():
    row1 = "{} | {} | {}".format(board[0], board[1], board[2]) 
    row2 = "{} | {} | {}".format(board[3], board[4], board[5])
    row3 = "{} | {} | {}".format(board[6], board[7], board[8])

    print()
    print(row1)
    print("-" * 9)
    print(row2)
    print("-" * 9)
    print(row3)
```

```python
# Check if the active player passed in as a parameter won the game
# Return True if they won and False if no win
#Param - Player - active player name

def check_win(player, icon):
    if (board[0] == icon and board[1] == icon and board[2] == icon) or \
       (board[3] == icon and board[4] == icon and board[5] == icon) or \
       (board[6] == icon and board[7] == icon and board[8] == icon) or \
       (board[0] == icon and board[3] == icon and board[6] == icon) or \
       (board[1] == icon and board[4] == icon and board[7] == icon) or \
       (board[2] == icon and board[5] == icon and board[8] == icon) or \
       (board[0] == icon and board[4] == icon and board[8] == icon) or \
       (board[2] == icon and board[4] == icon and board[6] == icon):

        print("{} has won!".format(player))
        return True
    else:
        return False
```

```python
#Used for the computer algorithm to determine the best possible move
def check_win_computer(icon, board):
    if (board[0] == icon and board[1] == icon and board[2] == icon) or \
       (board[3] == icon and board[4] == icon and board[5] == icon) or \
       (board[6] == icon and board[7] == icon and board[8] == icon) or \
       (board[0] == icon and board[3] == icon and board[6] == icon) or \
       (board[1] == icon and board[4] == icon and board[7] == icon) or \
       (board[2] == icon and board[5] == icon and board[8] == icon) or \
       (board[0] == icon and board[4] == icon and board[8] == icon) or \
       (board[2] == icon and board[4] == icon and board[6] == icon):

        return True
    else:
        return False
```

```python
#INITIALIZES all names and icons
#returns the player names and their respective icons
def intialize():
    player_num = int(input("How many players are playing 1 or 2? "))
    
    if player_num == 1:
        player1 = input("What is your name? ")
        player2 = "Computer"
        icon1 = input("Hi {}!. What icon do you want to be: X or O?: ".format(player1))

        if icon1 == "X":
            icon2 = "O"
        else:
            icon2 = "X"
    else:
        player1 = input("What is your name player 1? ").strip()
        player2 = input("What is your name player 2? ").strip()
        icon1 = input('Hi {} and {}! What icon do you want to be {}? X or O: '.format(player1, player2, player1)).strip()
        
        if icon1 == "X":
            icon2 = "O"
        else:
            icon2 = "X"
        
    return player1, player2, icon1, icon2
```

```python
#checks if there is a tie
def check_tie():
    if " " not in board:
        print("It's a tie!")
        return True
    else:
        return False
```

```python
#clears the board for a new game
def clear_board():
    board = [" " for i in range(9)]
    return board
```

```python
#computer algorithm for computer play
#will try to determine the best possible move
def computer_play(icon):
    for i in range(len(board)):
        temp = board
        if i == " ":
            temp[i] = icon
            if check_win_computer(icon, temp):
                board[i] = icon
                break
    while True:
        num  = random.randint(0,8)
        if board[num] == " ":
            board[num] = icon
            break
            
```

```python
#askes the player their move and makes the move
def move(icon):
    temp = int(input("Please enter your move between 1-9: ").strip()) - 1

    if board[temp] == " ":
        board[temp] = icon
    else:
        print("That space is already taken. Please pick again.")
        move(icon)
```

```python
#main function of the game
def play_game():
    print("Welcome to Tic-Tac-Toe! \n")
    board = clear_board()
    player1, player2, icon1, icon2 = intialize()
    

    while True:
        print("\n"+ ("-" * 30) + "\n")
        display_board()
        move(icon1)
        display_board()
        if check_win(player1, icon1):
            break
        elif check_tie():
            break
        
        if player2 == "Computer":
            computer_play(icon2)
            display_board()
            
            if check_win(player2, icon2):
                break
            elif check_tie():
                break
        
        else:
            move(icon2)
            display_board()
            if check_win(player2, icon2):
                break
            elif check_tie():
                break
```

```python
play_game()
```
